From the above we can see all three rows have one Residence State in common and also two are white in Race and the same two are in the same Residence County

We therefore group our data to see if we can find something usefull


```
Sra=data.groupby(['Sex','Race','ResidenceState','ResidenceCounty'])['Age'].median()
```

```
Sra
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print (Sra)
```

We find that a person of White Race,resides in Fairfiled County and in the Residence State of CT is roughly 41 years old if Female and 40 if Male

If the person is of mixed race (White and Hispanic),resides in HartFord and in the Residece State of CT the person is roughly 46 years if male and 48 if female



---

Our values are 56,41 and 39.
I'll choose Female for the white race and Male for the Hispanic, white and the corresponding values are closer to the median age

```
indicesF=[1430,3585]
indicesM=[3977]
data.loc[indicesF,'Sex']='Female'
data.loc[indicesM,'Sex']='Male'
```
