```
import requests
import json
headers = {
  'Content-Type': 'application/json',
  'x-api-key': 'insert-your-api-key-here'
}
lat = '37.7575756'
lon = '-122.5076424'
distance = '20mi'
datetime_ini = '2020-01-05T00:00:00.000Z'
datetime_end = '2020-01-06T00:00:00.000Z'
page = '1'
r = requests.get('https://api.crimeometer.com/v1/incidents/raw-data?lat='+lat+'&lon='+lon+'&distance='+distance+'&datetime_ini='+datetime_ini+'&datetime_end='+datetime_end+'&page='+page, headers=headers)
response = json.loads(r.text)
print(response)
```

```
incidentsList=response['incidents']
```

```
incidentsList
```

```
city_key=[]
incident_address=[]
incident_code=[]
incident_date=[]
incident_latitude=[]
incident_longitude=[]
incident_offense=[]
incident_offense_action=[]
incident_offense_code=[]
incident_offense_crime_against=[]
incident_offense_description=[]
incident_offense_detail_description=[]
incident_source_name=[]
incident_source_original_type=[]
length=len(incidentsList)
order=0

while order < length:
  cases=incidentsList[order]
  for key in cases.keys():
    if key == 'incident_code':
      values= cases[key]
      incident_code.append(values)
    elif key == 'incident_address':
      values= cases[key]
      incident_address.append(values)
    elif key == 'city_key':
      values= cases[key]
      city_key.append(values)
    elif key == 'incident_date':
      values=cases[key]
      incident_date.append(values)
    elif key == 'incident_latitude':
      values=cases[key]
      incident_latitude.append(values)
    elif key == 'incident_longitude':
      values = cases[key]
      incident_longitude.append(values)
    elif key == 'incident_offense':
      values= cases[key]
      incident_offense.append(values)
    elif key == 'incident_offense_action':
      values=cases[key]
      incident_offense_action.append(values)
    elif key == 'incident_offense_code':
      values = cases[key]
      incident_offense_code.append(values)
    elif key == 'incident_offense_crime_against':
      values = cases[key]
      incident_offense_crime_against.append(values)
    elif key == 'incident_offense_description':
      values = cases[key]
      incident_offense_description.append(values)
    elif key == 'incident_offense_detail_description':
      values= cases[key]
      incident_offense_detail_description.append(values)
    elif key == 'incident_source_name':
      values= cases[key]
      incident_source_name.append(values)
    elif key == 'incident_source_original_type':
      values=cases[key]
      incident_source_original_type.append(values)
    
```
